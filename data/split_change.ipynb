```python
train_set = image_ids[:int(len(image_ids)*90/100)]
val_set = image_ids[len(train_set):len(train_set) + int((len(image_ids) - len(train_set))/2)]
test_set = image_ids[len(train_set) + int((len(image_ids) - len(train_set))/2):]
```

```python
new_splits = {}

for elem in train_set:
    new_splits[elem] = 'train'
for elem in val_set:
    new_splits[elem] = 'val'
for elem in test_set:
    new_splits[elem] = 'test'
```

```python
import pandas as pd
from glob import glob

results_path_TEST = './results_TEST/'

batch_results_TEST = glob(results_path_TEST + '*.csv')
```

```python
dfs = []
for each_file in batch_results_TEST:
    df = pd.read_csv(each_file)
    dfs.append(df)
batch_results_TEST_df = pd.concat(dfs)
```

```python
res_df_vis = batch_results_TEST_df[['Input.Image',
       'Input.Paragraph', 'Input.Model',
       'Answer.objSalience_VISUAL', 'Answer.parCoherence_VISUAL',
       'Answer.sentStructure_VISUAL','Answer.wordChoice_VISUAL']]
res_df_vis = res_df_vis[res_df_vis['Input.Model'] == 'VISUAL']

res_df_lang = batch_results_TEST_df[['Input.Image',
       'Input.Paragraph', 'Input.Model',
       'Answer.objSalience_LANGUAGE', 'Answer.parCoherence_LANGUAGE',
       'Answer.sentStructure_LANGUAGE','Answer.wordChoice_LANGUAGE']]
res_df_lang = res_df_lang[res_df_lang['Input.Model'] == 'LANGUAGE']

res_df_vis_lang = batch_results_TEST_df[['Input.Image',
       'Input.Paragraph', 'Input.Model',
       'Answer.objSalience_VIS_LANG', 'Answer.parCoherence_VIS_LANG',
       'Answer.sentStructure_VIS_LANG','Answer.wordChoice_VIS_LANG']]
res_df_vis_lang = res_df_vis_lang[res_df_vis_lang['Input.Model'] == 'VIS_LANG']

res_df_vis_att = batch_results_TEST_df[['Input.Image',
       'Input.Paragraph', 'Input.Model',
       'Answer.objSalience_VISUAL_ATT', 'Answer.parCoherence_VISUAL_ATT',
       'Answer.sentStructure_VISUAL_ATT','Answer.wordChoice_VISUAL_ATT']]
res_df_vis_att = res_df_vis_att[res_df_vis_att['Input.Model'] == 'VISUAL_ATT']

res_df_lang_att = batch_results_TEST_df[['Input.Image',
       'Input.Paragraph', 'Input.Model',
       'Answer.objSalience_LANGUAGE_ATT', 'Answer.parCoherence_LANGUAGE_ATT',
       'Answer.sentStructure_LANGUAGE_ATT','Answer.wordChoice_LANGUAGE_ATT']]
res_df_lang_att = res_df_lang_att[res_df_lang_att['Input.Model'] == 'LANGUAGE_ATT']

res_df_vis_lang_att = batch_results_TEST_df[['Input.Image',
       'Input.Paragraph', 'Input.Model',
       'Answer.objSalience_VIS_LANG_ATT', 'Answer.parCoherence_VIS_LANG_ATT',
       'Answer.sentStructure_VIS_LANG_ATT','Answer.wordChoice_VIS_LANG_ATT']]
res_df_vis_lang_att = res_df_vis_lang_att[res_df_vis_lang_att['Input.Model'] == 'VIS_LANG_ATT']
```
